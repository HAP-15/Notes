```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def fair_roll(n):
    outcomes = [1,2,3,4,5,6]
    rng = np.random.default_rng()
    die_roll = rng.choice(outcomes, size = n)
    amount_list = []
    for value in outcomes:
        count = 0
        for num in die_roll:
            if num == value:
                count += 1           
        amount_list.append(count)   
    percentage_list = []
    for vals in amount_list:
        percentage_list.append(vals/n)
    plt.bar(outcomes, percentage_list)
    plt.show()
    
    plt.hist(die_roll, bins = 6, range = (1,7))
    plt.show()
    
    return percentage_list

def unfair_roll(n):
    outcomes = [1,2,3,4,5,6]
    rng = np.random.default_rng()
    die_roll = rng.choice(outcomes, size = n, p = np.array([.2,.1,.2,.1,.1,.3]))
    amount_list = []
    for value in outcomes:
        count = 0
        for num in die_roll:
            if num == value:
                count += 1           
        amount_list.append(count)   
    percentage_list = []
    for vals in amount_list:
        percentage_list.append(vals/n)
    
    plt.bar(outcomes, percentage_list)
    plt.show()

    plt.hist(die_roll, bins = 6, range = (1,7))
    plt.show()
    
    return percentage_list
```

```python
fair_roll(1000)
```

```python
unfair_roll(1000)
```

```python
def RandomWalk(num_steps, start_position):
    current_position = start_position
    steps = np.random.choice([0,1], num_steps)
    for step in (steps):
        if step == 1:
            current_position += 1
        elif step == 0:
            current_position -= 1 
    return current_position

def SimulateRandomWalk(num_sims, num_steps, start_position):
    results = []
    for sims in range(num_sims):
        a_sim = RandomWalk(num_steps, start_position)
        results.append([a_sim])
    return np.array(results)
```

```python
RandomWalk(10,0)
```

```python
SimulateRandomWalk(10,10,0)
```

```python
plt.hist(SimulateRandomWalk(10000,10,0))
plt.show()
```

```python
plt.hist(SimulateRandomWalk(10000,100,0))
plt.show()
```

```python
plt.hist(SimulateRandomWalk(10000,10000,0))
plt.show()
```
